```python
"""
https://fivethirtyeight.com/features/can-you-outwit-the-tax-collector/

The original Python script is in ./Documents/Python Files
"""
import math 
```

```python
#Define functions

def remove(l, m):
    """
    Returns a new list of elements in l not in m
    """
    
    newList = [] 
    
    for element in l:
        if element not in m:
            newList.append(element)
    
    return newList

def recursiveAddElement(combinations, l):
    """
    Creates all possible combinations of elements of list l
    """
    if len(combinations) == math.factorial(len(l)) and len(combinations[0]) == len(l):
        return combinations
    
    else: 
            
        newCombinations = []
        for combination in combinations:
            m = remove(l, combination)
            for element in m:
                newCombinations.append(combination + element) 
        
        #print(newCombinations) 
    
        return recursiveAddElement(newCombinations, l)
        

def happy(l):
    

    return int(l.index('y')) \
    + 10 * int(l.index('p')) \
    + 100 * int(l.index('p')) \
    + 1000 * int(l.index('a')) \
    + 10000 * int(l.index('h')) 


def holidays(l): 
    
    return int(l.index('s')) \
    + 10 * int(l.index('y')) \
    + 100 * int(l.index('a')) \
    + 1000 * int(l.index('d')) \
    + 10000 * int(l.index('i')) \
    + 100000 * int(l.index('l')) \
    + 1000000 * int(l.index('o')) \
    + 10000000 * int(l.index('h')) 
    
    
def hohohoho(l):
    
    return int(l.index('o')) \
    + 10 * int(l.index('h')) \
    + 100 * int(l.index('o')) \
    + 1000 * int(l.index('h')) \
    + 10000 * int(l.index('o')) \
    + 100000 * int(l.index('h')) \
    + 1000000 * int(l.index('o')) \
    + 10000000 * int(l.index('h')) 

def happyHolidaysAdd(l):
    """
    Evaluates if the list is a solution to the happy holidays question, 
    using index as the digit
    """
    
    return (happy(l) + holidays(l) == hohohoho(l)) 
```

```python
#main() 
l = ['h', 'a', 'p', 'y', 'o', 'l', 'i', 'd', 's', ' ']

m = [i for i in l]

combos = recursiveAddElement(m, l)

print('Done computing combos') 

for i in combos:
    if happyHolidaysAdd(i) == True:
        print(i, '\n', 'Happy: ', happy(i), '\t', 'Holidays: ', holidays(i), '\t', 'Hohohoho: ', hohohoho(l))
        print('\n') 
```
